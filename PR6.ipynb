7. Створити зведену таблицю, щоб класифікувати дані за «роком», і вивести
середні значення показника щастя для кожного року.

```python
pivotY = pd.pivot_table(
    df,
    index='Year',
    values='Happiness Score',
    aggfunc='mean'
)
print(pivotY)
```

8. Повторити попередній крок для стовпця «Регіон».

```python
pivotR = pd.pivot_table(
    df,
    index='Region',
    values='Happiness Score',
    aggfunc='mean'
)
print(pivotR)
```

9. Вивести середні значення показника щастя для кожного року та регіону (у 1 зведеній таблиці).

```python
pivotRY = pd.pivot_table(
    df,
    index=['Region','Year'],
    values='Happiness Score',
    aggfunc='mean'
)
print(pivotRY)
```

10. Змінити попереднє рішення для отримання такого результату

```python
pivotRY = pd.pivot_table(
    df,
    values='Happiness Score',
    index='Region',
    columns='Year',
    aggfunc='mean'
)
pivotRY
```

11. Створити зведену таблицю на основі регіону, щоб отримати такий
результат

```python
pivotR = pd.pivot_table(
    df,
    values='Happiness Score',
    index='Region',
    aggfunc=['mean', 'median', 'min', 'max', 'std'],
)
pivotR
```

12. Змінити попереднє рішення для отримання такого результату:

```python
pivotR = pd.pivot_table(
    df,
    values='Happiness Score',
    index='Region',
    aggfunc=['mean', 'median', 'min', 'max', 'std'],
    columns='Year'
)
pivotR
```
