```python
import pandas as pd

pop = pd.read_csv('state-pop_my.csv', sep=';')
areas = pd.read_csv('state-areas_my.csv', sep=';')
abbrevs = pd.read_csv('state-abbrevs_my.csv', sep=';')
```

6. За допомогою методу filter() виведіть 10 перших груп, у яких стандартне відхилення
