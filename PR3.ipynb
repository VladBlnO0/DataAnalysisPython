### Завдання 1

1. Створіть таблицю для України розміром N1×22 (N1 крамниця × 12 місяців 2023 року і перші 10
місяців 2024 року). Заповніть її цілими значеннями у діапазоні від 120 до 420 включно.

```python
import numpy as np
rng = np.random.default_rng(42)

ukraine = np.array(rng.integers(120, 420, size=(11, 22), endpoint=True))
ukraine
```

2. Створіть таблицю для Польщі розміром N2×22 (N2 крамниця × 12 місяців 2023 року і перші 10
місяців 2024 року). Заповніть її цілими значеннями у діапазоні від 90 до 300 включно.

```python
poland = np.array(rng.integers(90, 300, size=(9, 22), endpoint=True))
poland
```

3. Присвойте кожному супермаркету реалістичну назву. Місяці позначте українськими назвами.

```python
ukraine_stores = [f"Україна_{i+1}" for i in range(ukraine.shape[0])]
poland_stores = [f"Польща_{i+1}" for i in range(poland.shape[0])]
store_names = np.concatenate((ukraine_stores, poland_stores))

months = [
    "Січень_23", "Лютий_23", "Березень_23", "Квітень_23", "Травень_23", "Червень_23",
    "Липень_23", "Серпень_23", "Вересень_23", "Жовтень_23", "Листопад_23", "Грудень_23",
    "Січень_24", "Лютий_24", "Березень_24", "Квітень_24", "Травень_24",
    "Червень_24", "Липень_24", "Серпень_24", "Вересень_24", "Жовтень_24"
]
```

### Завдання 2

Створіть одну спільну таблицю такого вигляду:
* рядки – усі крамниці обох країн (усього 20 рядків);
* стовпчики – місяці для аналізу: червень, липень, серпень, вересень 2023, лютий, березень, квітень, травень 2024.
<p> Для цього візьміть зі створених у п.1 таблиць відповідні колонки і поєднайте їх по рядках у зазначеному
порядку місяців.
Далі саме ця матриця використовуватиметься для аналізу.

```python
months_needed = ["Червень_23", "Липень_23", "Серпень_23", "Вересень_23",
                  "Лютий_24", "Березень_24", "Квітень_24", "Травень_24"]
columns = [months.index(m) for m in months_needed]

ukraine_sel = ukraine[:, columns]
poland_sel = poland[:, columns]

combined = np.vstack((ukraine_sel, poland_sel))
combined
```

### Завдання 3

1. Для кожного з М місяців обчисліть: суму, найбільше та найменше значення обсягів експорту
(по всіх крамницях).

```python
sum_by_month = combined.sum(0)
sum_by_month
```

```python
max_by_month = combined.max(axis=0)
max_by_month
```

```python
min_by_month = combined.min(axis=0)
min_by_month
```

2. Для кожного обʼєкта обчисліть середній обсяг експорту за ці М місяців.

```python
mean_by_month = combined.mean(axis=1)
for i, name in enumerate(store_names):
    print(f"{name}: {mean_by_month[i]}")
```

3. Результати подайте окремо:
* вектор із M сум;
* вектор із M найбільших значень та відповідних їм місяців і обʼєктів;
* вектор із M найменших значень та відповідних їм місяців і обʼєктів;
* вектор із N1+N2 середнього (по кожному обʼєкту).

```python
sum_by_month = combined.sum(axis=0)
sum_by_month
```

```python
max_values = combined.max(axis=0)
max_indices = combined.argmax(axis=0)

for m_idx, (val, s_idx) in enumerate(zip(max_values, max_indices)):
    print(f"{months_needed[m_idx]}: {store_names[s_idx]} - {val}")
```

```python
min_values = combined.min(axis=0)
min_indices = combined.argmin(axis=0)

for m_idx, (val, s_idx) in enumerate(zip(min_values, min_indices)):
    print(f"{months_needed[m_idx]}: {store_names[s_idx]} - {val}")
```
